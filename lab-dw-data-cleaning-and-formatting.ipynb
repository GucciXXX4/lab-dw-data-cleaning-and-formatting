## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)

print("Before cleaning:")
display(df.head())

#check the column names in the DataFrame

print("\nActual column names in the dataset:")
print(df.columns.tolist())

#cleaning invalid values
#standardize gender

if 'Gender' in df.columns:
    df['Gender'] = df['Gender'].str.strip().str.lower()
    df['Gender'] = df['Gender'].replace({
        'f': 'F', 'female': 'F', 'femal': 'F',
        'm': 'M', 'male': 'M'
    })
    
elif 'gender' in df.columns:
    df['gender'] = df['gender'].str.strip().str.lower()
    df['gender'] = df['gender'].replace({
        'f': 'F', 'female': 'F', 'femal': 'F',
        'm': 'M', 'male': 'M'
    })

#replace state abbreviations with full names
state_col = 'state' if 'state' in df.columns else 'State' if 'State' in df.columns else None
if state_col:
    df[state_col] = df[state_col].replace({
        'AZ': 'Arizona',
        'Cali': 'California',
        'WA': 'Washington'
    })

#clean education values
edu_col = 'education' if 'education' in df.columns else 'Education' if 'Education' in df.columns else None
if edu_col:
    df[edu_col] = df[edu_col].replace({
        'Bachelors': 'Bachelor'
    })

#remove % from customer lifetime value and convert to float
clv_col = 'customer_lifetime_value' if 'customer_lifetime_value' in df.columns else 'Customer Lifetime Value' if 'Customer Lifetime Value' in df.columns else None
if clv_col:
    df[clv_col] = df[clv_col].astype(str).str.replace('%', '', regex=True)
    df[clv_col] = pd.to_numeric(df[clv_col], errors='coerce')

#standardize vehicle class
vc_col = 'vehicle_class' if 'vehicle_class' in df.columns else 'Vehicle Class' if 'Vehicle Class' in df.columns else None
if vc_col:
    df[vc_col] = df[vc_col].replace({
        'Sports Car': 'Luxury',
        'Luxury SUV': 'Luxury',
        'Luxury Car': 'Luxury'
    })

print("\nAfter cleaning:")

#display only the columns we've cleaned
columns_to_display = []
for col in [('gender', 'Gender'), ('state', 'State'), ('education', 'Education'), 
           ('customer_lifetime_value', 'Customer Lifetime Value'), ('vehicle_class', 'Vehicle Class')]:
    if col[0] in df.columns:
        columns_to_display.append(col[0])
    elif col[1] in df.columns:
        columns_to_display.append(col[1])

display(df[columns_to_display].head(10))

```

## Exercise 3: Formatting data types

The data types of many columns in the dataset appear to be incorrect. This could impact the accuracy of our analysis. To ensure accurate analysis, we need to correct the data types of these columns. Please update the data types of the columns as appropriate.

It is important to note that this exercise does not involve handling null values (NaN or None).

*Hint*:
- *Customer lifetime value should be numeric*
- *Number of open complaints has an incorrect format. Look at the different values it takes with `unique()` and take the middle value. As an example, 1/5/00 should be 5. Number of open complaints is a string - remember you can use `split()` to deal with it and take the number you need. Finally, since it should be numeric, cast the column to be in its proper type.*

```python
# Your code here
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)

#ensure customer lifetime value is numeric

df['Customer Lifetime Value'] = (
    df['Customer Lifetime Value']
    .astype(str)
    .str.replace('%', '', regex=True)
)
df['Customer Lifetime Value'] = pd.to_numeric(df['Customer Lifetime Value'], errors='coerce')

#fix number of open complaints
print("Unique values before cleaning:")
print(df['Number of Open Complaints'].unique())

# extract the middle number if the format is like 1/5/00
def extract_middle(val):
    try:
        parts = str(val).split('/')
        if len(parts) >= 2:
            return int(parts[1])
        else:
            return int(val)
    except:
        return None

df['Number of Open Complaints'] = df['Number of Open Complaints'].apply(extract_middle)

#convert to numeric
df['Number of Open Complaints'] = pd.to_numeric(df['Number of Open Complaints'], errors='coerce')

print("\nData types after cleaning:")
print(df.dtypes)

print("\ncleaned columns:")
print(df[['Customer Lifetime Value', 'Number of Open Complaints']].head(10))
```

## Exercise 4: Dealing with Null values

Identify any columns with null or missing values. Identify how many null values each column has. You can use the `isnull()` function in pandas to find columns with null values.

Decide on a strategy for handling the null values. There are several options, including:

- Drop the rows or columns with null values
- Fill the null values with a specific value (such as the column mean or median for numerical variables, and mode for categorical variables)
- Fill the null values with the previous or next value in the column
- Fill the null values based on a more complex algorithm or model (note: we haven't covered this yet)

Implement your chosen strategy to handle the null values. You can use the `fillna()` function in pandas to fill null values or `dropna()` function to drop null values.

Verify that your strategy has successfully handled the null values. You can use the `isnull()` function again to check if there are still null values in the dataset.

Remember to document your process and explain your reasoning for choosing a particular strategy for handling null values.

After formatting data types, as a last step, convert all the numeric variables to integers.

```python
# Your code here
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)

#check for null values

print("Null values per column before cleaning:")
print(df.isnull().sum())

#handle null values

for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].median(), inplace=True)

#check to verify  
print("\nNull values per column after cleaning:")
print(df.isnull().sum())

#convert all numeric variables to integers
#identify numeric columns

numeric_cols = df.select_dtypes(include=[np.number]).columns

#convert to integers
df[numeric_cols] = df[numeric_cols].astype(int)

print("\nData types after conversion:")
print(df.dtypes)


print("\nPreview of cleaned data:")
print(df.head())
```

```python
# Your code here
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
df = pd.read_csv(url)

#check for duplicates
print("Number of duplicate rows before cleaning:", df.duplicated().sum())

#drop all duplicates and keep only the first occurrence
df = df.drop_duplicates(keep='first')

#reset index
df = df.reset_index(drop=True)

#check to verify
print("Number of duplicate rows after cleaning:", df.duplicated().sum())
print("Shape after cleaning:", df.shape)

#save cleaned dataset
df.to_csv("file1_cleaned.csv", index=False)
print("Cleaned dataset saved as file1_cleaned.csv")
```
